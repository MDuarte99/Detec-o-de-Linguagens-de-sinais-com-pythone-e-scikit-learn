# 1. Importe das biblíotecas necessárias

```python
import os
import cv2
import string
```

# 2. Cria o diretório necessário para salvar imagens capturadas

```python
DATA_DIR = './data'
if not os.path.exists(DATA_DIR):
    os.makedirs(DATA_DIR)
```

# 3. Captura a letras em Libras para essse exemplo capturaremos:

```python
number_of_classes = 3 # números de letras a serem capturadas
dataset_size = 100    # número de frames

cap = cv2.VideoCapture(1) # Câmera está alocado no valor "1", procure a sua
for j in range(number_of_classes):
    if not os.path.exists(os.path.join(DATA_DIR, str(j))):
        os.makedirs(os.path.join(DATA_DIR, str(j)))

    print('Coletando dados para a classe {}'.format(j))

    done = False
    while True:
        ret, frame = cap.read()
        # Dimencionamento da imagem
        cv2.putText(frame, 'Pronto? Aperte "Q" ! :)', (100, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3,
                    cv2.LINE_AA)
        cv2.imshow('frame', frame)
        if cv2.waitKey(25) == ord('q'):
            break

    counter = 0
    while counter < dataset_size:
        ret, frame = cap.read()
        cv2.imshow('frame', frame)
        cv2.waitKey(25)
        cv2.imwrite(os.path.join(DATA_DIR, str(j), '{}.jpg'.format(counter)), frame)

        counter += 1

cap.release()
cv2.destroyAllWindows()
```
